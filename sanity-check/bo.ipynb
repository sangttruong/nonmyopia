```python
import numpy as np

mean = np.array([2, 3, 2])

cov = np.array([
    [1, 2, 0],
    [2, 5, 0],
    [0, 0, 1]
])

y = np.random.multivariate_normal(mean, cov, 1000)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(4, 4))
plt.scatter(y[:, 0], y[:, 1])
plt.xlim(-10, 10)
plt.ylim(-10, 10)
```

```python
plt.hist(y[:, 0], density=True, bins=50, color='red', alpha=0.5)
plt.hist(y[:, 2], density=True, bins=50, color='blue', alpha=0.5)
plt.show()
```
